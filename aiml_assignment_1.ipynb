```python
#Part-1:Implement Basic Data Structures using Numpy, Pandas
import numpy as np
#Lists
#Arrays
#Type
list1=[1,2,3,4,5]
print(list1)
print(type(list1))
list2=[11,12,13,14,15]
print(list2)
print(type(list2))
arr1=np.array(list1)
print(arr1)
print(type(arr1))
arr2=np.array(list2)
print(arr2)
print(type(arr2))
#Mathematical operations
Add=list1+list2
print("ADDITION OF LISTS:",Add)
Multi=arr1*arr2
print("MULTIPLICATION OF LISTS:",Multi)
Sub=arr1-arr2
print("SUBTRACTION OF LISTS:",Sub)
Div=arr1/arr2
print("DIVISION OF LISTS:",Div)
power=arr1**arr2
print("POWER OF LISTS:",power)
#tuple
tuple1=("moniter","mouse","keyboard")
print(tuple1)
print(type(tuple1))
tuple2=(100,200,300)
print(tuple2)
print(type(tuple2))
#sin
sin=np.sin(arr1)
print("SIN VALUE:",sin)
#log
log=np.log(arr1)
print("LOG VALUE:",log)
#log2
log2=np.log2(arr1)
print("LOG2 VALUE:",log2)
#exp
exp=np.exp(arr1)
print("EXP VALUE:",exp)



```

```python
#part-2
import pandas as pd # Import pandas as pd
import matplotlib.pyplot as plt
#DATA
people=['sam','chay','shobi','nag']
ages=[25,30,35,40]
#removed the extra element to match the length of the lists
heights=[145,151,165,173]
weights=[45,55,65,75]
#create a dataframe
data={'people':people,'ages':ages,'heights':heights,'weights':weights}
df=pd.DataFrame(data) # Use pd to create the DataFrame
#print the dataframe
print(df)
#scatter plot:height vs weight
plt.figure(figsize=(8,6))
# Corrected column names here to remove '(cm)' and '(kg)'
plt.scatter(df['heights'],df['weights'],color='blue',marker='o')
plt.title('Scatter plot of height vs weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(True)
plt.show()
#bar chart:heights and weights of individuals
plt.figure(figsize=(8,6))
bar_width=0.4
index=range(len(df['people'])) # Changed 'Name' to 'people'
plt.bar(index,df['heights'],bar_width,label='Height(cm)',color='blue') # Changed 'Height(cm)' to 'heights'
plt.bar([i+bar_width for i in index],df['weights'],bar_width,label='Weight(kg)',color='green') # Changed 'Weight(kg)' to 'weights'
plt.title('Bar chart of heights and weights of individuals')
plt.xlabel('Individuals')
plt.ylabel('Measurements')
plt.xticks([i+bar_width/2 for i in index],df['people']) # Changed 'Name' to 'people'
plt.legend()
plt.grid(True)
plt.show()

#Histogram:distribution of age,height,and weight
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.hist(df['ages'],bins=5,color='purple',edgecolor='black') # Changed 'Age' to 'ages'
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.subplot(1,3,2)
plt.hist(df['heights'],bins=4,color='orange',edgecolor='black') # Changed 'Height(cm)' to 'heights'
plt.title('Distribution of Height')
plt.xlabel('Height(cm)')
plt.ylabel('Frequency')
plt.subplot(1,3,3)
plt.hist(df['weights'],bins=4,color='red',edgecolor='black') # Changed 'Weight(kg)' to 'weights'
plt.title('Distribution of Weight')
plt.xlabel('Weight(kg)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
